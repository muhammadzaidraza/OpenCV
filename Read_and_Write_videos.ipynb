```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
cap = cv2.VideoCapture('./drive/MyDrive/video/videos/test.mp4')
if (cap.isOpened() == False):
  print('error opening video file')
```

```python
#reading and displaying one frame

ret , frame = cap.read()
plt.imshow(frame[...,::-1])
```

```python
#writing video file

width = int(cap.get(3))
height = int(cap.get(4))

avi_out = cv2.VideoWriter("output.avi" , cv2.VideoWriter_fourcc('M','J','p','G') , 10 , (width,height));

mp4_out = cv2.VideoWriter("output.mp4" , cv2.VideoWriter_fourcc(*'XVID') , 10 , (width,height))

#read untill video is completed

while(cap.isOpened()):
  #capture frame by frame
  ret , frame = cap.read()

  if ret == True:

    #write the frame to output file
    avi_out.write(frame)
    mp4_out.write(frame)
  
  #break the loop
  else:
    break

#realeasing resources when utilized
cap.release()
avi_out.release()
mp4_out.release()
```
