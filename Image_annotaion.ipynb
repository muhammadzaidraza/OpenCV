```python
from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
from PIL import Image
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0,9.0)
```

```python
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",0)
#print(img_color)
plt.imshow(img_color,cmap="gray")
```

```python
# drawing lines on images
annotate = cv2.line(img_color, (100,100),(300,300),(0,0,0), thickness=10, lineType=cv2.LINE_AA)
plt.imshow(annotate)
```

```python
# drawing circle on images
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",0)
annotate = cv2.circle(img_color, (360,260), 50 ,(0,0,0), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(annotate)
```

```python
# drawing rectangle on images
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",0)
annotate = cv2.rectangle(img_color, (300,200), (400,300) ,(255,255,255), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(annotate)
```

```python
# putting text on images
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",0)
annotate = cv2.rectangle(img_color, (300,200), (400,300) ,(255,255,255), thickness=2, lineType=cv2.LINE_AA)
cv2.putText(annotate, "bird", (300,200), cv2.FONT_HERSHEY_PLAIN , 2 , (0,0,0) , 2 , cv2.LINE_AA )
plt.imshow(annotate)
```
