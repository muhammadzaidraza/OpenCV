```python
from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
from PIL import Image
```

```python
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",1)
img_color = cv2.cvtColor(img_color , cv2.COLOR_BGR2RGB)
#print(img_color)
plt.imshow(img_color)
```

```python
# creating brighter and darker image
matrix = np.ones(img_color.shape , dtype="uint8")*50
brighter_img = cv2.add(img_color , matrix)
darker_img = cv2.subtract(img_color , matrix)

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(darker_img);plt.title("Darker")
plt.subplot(132);plt.imshow(img_color);plt.title("Original")
plt.subplot(133);plt.imshow(brighter_img);plt.title("Brighter")
```

```python
#image contrast
matrix1 = np.ones(img_color.shape, dtype="uint8") * 0.8
matrix2 = np.ones(img_color.shape, dtype="uint8") * 1.2

darker_img =np.uint8(cv2.multiply(np.float64(img_color),matrix1)) #as matrix contain float value therfor we first convert image into float values
brighter_img = np.uint8(cv2.multiply(np.float64(img_color),matrix2))

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(darker_img);plt.title("Darker")
plt.subplot(132);plt.imshow(img_color);plt.title("Original")
plt.subplot(133);plt.imshow(brighter_img);plt.title("Brighter")

#in brighter image overflow occurs which cause some undesired colors
```

```python
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",0)
img_color = cv2.cvtColor(img_color , cv2.COLOR_BGR2RGB)

#applying threshold
thresh = 100 #above 100 all values will be 255 and below 100 all values will be 0
maxval = 255 # maximum valu after threshld is aplied
type = cv2.THRESH_BINARY
reteval, img_thresh = cv2.threshold(img_color, thresh, maxval, type )

plt.figure(figsize=[18,5])
plt.subplot(121);plt.imshow(img_thresh);plt.title("Threshold image")
plt.subplot(122);plt.imshow(img_color);plt.title("Original image")
```
