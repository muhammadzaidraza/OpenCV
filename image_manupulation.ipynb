```python
from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
from PIL import Image
```

```python
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",1)
img_color = cv2.cvtColor(img_color,cv2.COLOR_BGR2RGB)
print(img_color)
plt.imshow(img_color,cmap="gray")
```

```python
#accessing the specific pixels of the image
print(img_color[0,0])
print(img_color[399,599])
```

```python
#modifying image pixels
img_copy = img_color.copy()
img_copy[50:100,50:100] = 200
print(img_copy)
plt.imshow(img_copy)
```

```python
#cropping image
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",1)
img_color = cv2.cvtColor(img_color,cv2.COLOR_BGR2RGB)
cropped_region = img_color[200:400,100:300]
plt.imshow(cropped_region)
```

```python
double_size = cv2.resize(cropped_region , None , fx = 2 ,fy = 2)
plt.imshow(double_size)
```

```python
fix_size =  cv2.resize(img_color,dsize=(64,64),interpolation = cv2.INTER_LANCZOS4)
plt.imshow(fix_size)
```

```python
#resizing while maintaining the aspect ratio
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",1)
img_color = cv2.cvtColor(img_color,cv2.COLOR_BGR2RGB)
desired_width = 64
aspect_ratio = desired_width/img_color.shape[1]
height = int(img_color.shape[0]*aspect_ratio)
dim = (desired_width,height)
img = cv2.resize(img_color , dsize = dim , interpolation = cv2.INTER_AREA)
plt.imshow(img)
```

```python
#image flipping
img_color = cv2.imread("./gdrive/MyDrive/1.jpg",0)

img_hz = cv2.flip(img_color,1) 
img_vt = cv2.flip(img_color,0)
img_bt = cv2.flip(img_color,-1)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(img_hz, cmap='gray');plt.title("horizontal flip");
plt.subplot(142);plt.imshow(img_vt,cmap='gray');plt.title("vertical flip");
plt.subplot(143);plt.imshow(img_bt,cmap='gray');plt.title("both flip");
plt.subplot(144);plt.imshow(img_color,cmap='gray');plt.title("original");
```
