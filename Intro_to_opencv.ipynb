```python
#reading image and spliting image
img = cv2.imread("./gdrive/MyDrive/1.jpg",1)
b,g,r = cv2.split(img)

#creating subplots and representing each channel
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r, cmap='gray');plt.title("red channel");
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("green channel");
plt.subplot(143);plt.imshow(b,cmap='gray');plt.title("blue channel");

#merge the channels into rgb format
imgmerged = cv2.merge((r,g,b))

#displaying image
plt.subplot(144);plt.imshow(imgmerged);plt.title("merged image");
```

```python
# converting image colors
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(img_hsv)

#creating subplots and representing each channel
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h);plt.title("h channel");
plt.subplot(142);plt.imshow(s);plt.title("s channel");
plt.subplot(143);plt.imshow(v);plt.title("v channel");
plt.subplot(144);plt.imshow(cv2.merge((h,s,v)));plt.title("merged image");
```
